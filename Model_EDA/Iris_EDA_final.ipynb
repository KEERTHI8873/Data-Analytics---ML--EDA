```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/iris_extended.csv")
print(" Dataset Loaded Successfully\n")
```

```python
print(" First 5 Rows of Dataset:\n", df.head())
print("\n Dataset Shape:", df.shape)
print("\n Dataset Info:")
print(df.info())
print("\n Missing Values:\n", df.isnull().sum())
print("\n Statistical Summary:\n", df.describe())
```

```python
if 'species' in df.columns:
    target_col = 'species'
elif 'class' in df.columns:
    target_col = 'class'
else:
    raise ValueError("No target column found ('species' or 'class').")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=df[target_col], palette="Set2")
plt.title("Target Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
import math

numeric_cols = df.select_dtypes(include='number').columns
n = len(numeric_cols)
rows = math.ceil(n / 2)  # 2 plots per row

plt.figure(figsize=(12, rows * 4))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(rows, 2, i)
    sns.boxplot(x=target_col, y=col, data=df, palette="Set3")
    plt.title(f"Boxplot of {col} by Class")

plt.tight_layout()
plt.show()
```

```python
df.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle("Feature Histograms", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=df.columns[0], y=df.columns[1],
    hue=target_col, size=df.columns[2],
    sizes=(20, 200), alpha=0.7, palette="viridis", data=df
)
plt.title(f"Scatter Plot: {df.columns[0]} vs {df.columns[1]}")
plt.show()
```
